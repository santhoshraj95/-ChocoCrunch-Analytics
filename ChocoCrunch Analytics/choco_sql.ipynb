## SQL Queries

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('SChoco.db')

# ---------- product_info Queries ----------

# 1. Count products per brand
q1 = pd.read_sql_query("SELECT brand, COUNT(*) AS product_count FROM product_info GROUP BY brand ORDER BY product_count DESC;", conn)

# 2. Count unique products per brand
q2 = pd.read_sql_query("SELECT brand, COUNT(DISTINCT product_name) AS unique_products FROM product_info GROUP BY brand ORDER BY unique_products DESC;", conn)

# 3. Top 5 brands by product count
q3 = pd.read_sql_query("SELECT brand, COUNT(*) AS product_count FROM product_info GROUP BY brand ORDER BY product_count DESC LIMIT 5;", conn)

# 4. Products with missing product name
q4 = pd.read_sql_query("SELECT * FROM product_info WHERE product_name IS NULL OR TRIM(product_name) = '';", conn)

# 5. Number of unique brands
q5 = pd.read_sql_query("SELECT COUNT(DISTINCT brand) AS unique_brand_count FROM product_info;", conn)

# 6. Products with code starting with '3'
q6 = pd.read_sql_query("SELECT * FROM product_info WHERE product_code LIKE '3%';", conn)

# ---------- nutrient_info Queries ----------

# 7. Top 10 products with highest energy-kcal_value
q7 = pd.read_sql_query('SELECT product_code, "energy-kcal_value" FROM nutrient_info ORDER BY "energy-kcal_value" DESC LIMIT 10;', conn)

# 8. Average sugars_value per nova-group
q8 = pd.read_sql_query('SELECT "nova-group", AVG(sugars_value) AS avg_sugars FROM nutrient_info GROUP BY "nova-group";', conn)

# 9. Count products with fat_value > 20g
q9 = pd.read_sql_query("SELECT COUNT(*) AS high_fat_products FROM nutrient_info WHERE fat_value > 20;", conn)

# 10. Average carbohydrates_value per product
q10 = pd.read_sql_query("SELECT AVG(carbohydrates_value) AS avg_carbohydrates FROM nutrient_info;", conn)

# 11. Products with sodium_value > 1g
q11 = pd.read_sql_query("SELECT * FROM nutrient_info WHERE sodium_value > 1;", conn)

# 12. Count products with non-zero fruits-vegetables-nuts content
q12 = pd.read_sql_query('SELECT COUNT(*) AS products_with_fvn FROM nutrient_info WHERE "fruits-vegetables-nuts-estimate-from-ingredients_100g" > 0;', conn)

# 13. Products with energy-kcal_value > 500
q13 = pd.read_sql_query('SELECT * FROM nutrient_info WHERE "energy-kcal_value" > 500;', conn)

# ---------- derived_metrics Queries ----------

# 14. Count products per calorie_category
q14 = pd.read_sql_query('SELECT calorie_category, COUNT(*) AS product_count FROM derived_metrics GROUP BY calorie_category;', conn)

# 15. Count of High Sugar products
q15 = pd.read_sql_query("SELECT COUNT(*) AS high_sugar_count FROM derived_metrics WHERE sugar_category = 'High Sugar';", conn)

# 16. Average sugar_to_carb_ratio for High Calorie products
q16 = pd.read_sql_query("SELECT AVG(sugar_to_carb_ratio) AS avg_ratio FROM derived_metrics WHERE calorie_category = 'High Calorie';", conn)

# 17. Products that are both High Calorie and High Sugar
q17 = pd.read_sql_query("SELECT * FROM derived_metrics WHERE calorie_category = 'High Calorie' AND sugar_category = 'High Sugar';", conn)

# 18. Number of products marked as ultra-processed
q18 = pd.read_sql_query("SELECT COUNT(*) AS ultra_processed_count FROM derived_metrics WHERE is_ultra_processed = 1;", conn)

# 19. Products with sugar_to_carb_ratio > 0.7
q19 = pd.read_sql_query("SELECT * FROM derived_metrics WHERE sugar_to_carb_ratio > 0.7;", conn)

# 20. Average sugar_to_carb_ratio per calorie_category
q20 = pd.read_sql_query("SELECT calorie_category, AVG(sugar_to_carb_ratio) AS avg_ratio FROM derived_metrics GROUP BY calorie_category;", conn)

# ---------- Join Queries ----------

# 21. Top 5 brands with most High Calorie products
q21 = pd.read_sql_query('''
    SELECT pi.brand, COUNT(*) AS high_calorie_count
    FROM derived_metrics dm
    JOIN product_info pi ON dm.product_code = pi.product_code
    WHERE dm.calorie_category = 'High Calorie'
    GROUP BY pi.brand
    ORDER BY high_calorie_count DESC
    LIMIT 5;
''', conn)

# 22. Average energy-kcal_value for each calorie_category
q22 = pd.read_sql_query('''
    SELECT dm.calorie_category, AVG(ni."energy-kcal_value") AS avg_energy_kcal
    FROM derived_metrics dm
    JOIN nutrient_info ni ON dm.product_code = ni.product_code
    GROUP BY dm.calorie_category;
''', conn)

# 23. Count of ultra-processed products per brand
q23 = pd.read_sql_query('''
    SELECT pi.brand, COUNT(*) AS ultra_processed_count
    FROM derived_metrics dm
    JOIN product_info pi ON dm.product_code = pi.product_code
    WHERE dm.is_ultra_processed = 1
    GROUP BY pi.brand
    ORDER BY ultra_processed_count DESC;
''', conn)

# 24. Products with High Sugar and High Calorie along with brand
q24 = pd.read_sql_query('''
    SELECT pi.product_name, pi.brand
    FROM derived_metrics dm
    JOIN product_info pi ON dm.product_code = pi.product_code
    WHERE dm.calorie_category = 'High Calorie' AND dm.sugar_category = 'High Sugar';
''', conn)

# 25. Average sugar content per brand for ultra-processed products
q25 = pd.read_sql_query('''
    SELECT pi.brand, AVG(ni.sugars_value) AS avg_sugars
    FROM derived_metrics dm
    JOIN product_info pi ON dm.product_code = pi.product_code
    JOIN nutrient_info ni ON dm.product_code = ni.product_code
    WHERE dm.is_ultra_processed = 1
    GROUP BY pi.brand
    ORDER BY avg_sugars DESC;
''', conn)

# 26. Number of products with fruits/vegetables/nuts content in each calorie_category
q26 = pd.read_sql_query('''
    SELECT dm.calorie_category, COUNT(*) AS product_count
    FROM derived_metrics dm
    JOIN nutrient_info ni ON dm.product_code = ni.product_code
    WHERE ni."fruits-vegetables-nuts-estimate-from-ingredients_100g" > 0
    GROUP BY dm.calorie_category;
''', conn)

# 27. Top 5 products by sugar_to_carb_ratio with their calorie and sugar category
q27 = pd.read_sql_query('''
    SELECT product_code, sugar_to_carb_ratio, calorie_category, sugar_category
    FROM derived_metrics
    ORDER BY sugar_to_carb_ratio DESC
    LIMIT 5;
''', conn)

# Example: print one result
pd.DataFrame(q1).head()# You can change q1 to q2, q3, ... q27 to see others

# Close connection when done
#conn.close()
```
